```python
from ast import literal_eval
data['lyrics_emo'] = data['lyrics_emo'].apply(literal_eval) #문자열 표현식으로 인식 변환
```

```python
# 주요 감정 추출
def main_emotion(d):
    k = v = ''
    if d:
        d = {k: v for k, v in d.items() if isinstance(v, (int, float))}
        if d:
            k = max(d, key=d.get)
            v = d[k]
    return pd.Series({'emotion': k, 'value': v})

data = pd.concat([data, data['lyrics_emo'].apply(main_emotion)], axis=1)
```

```python
data.drop(['lyrics', 'value', 'label'], axis=1, inplace=True)
# 필요 없어보이는 열들은 제거 했습니다 label
```
