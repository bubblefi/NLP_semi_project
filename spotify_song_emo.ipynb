```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import nltk
# # nltk.download('wordnet')

from ast import literal_eval
```

```python
emo=pd.read_csv('spotify_emo.csv')
```

```python
emo.drop(['artist', 'lem','cleaned_lyrics', 'emotion'], axis=1, inplace=True)
```

```python
emo['lyrics_emo'] = emo['lyrics_emo'].apply(literal_eval) #문자열 표현식으로 인식 변환
```

```python
# 참고 주소 딕셔너리의 키를 어떻게 열로? 
# How to Add Dictionary Keys and Values as Pandas Columns?
# https://www.datasciencelearner.com/how-to-add-dictionary-keys-and-values-as-pandas-columns/
```

```python
rows=[]
df = emo['lyrics_emo']
for song, row in zip(emo['song'], df):
    row['song'] = song
    rows.append(row)
    
data = pd.DataFrame(rows)
data
```

```python
data.columns
```

```python
# 데이터프레임 컬럼 순서 변경 - 'song'열이 맨 앞으로 오게 했습니다-
data=data[['song','disgust', 'negative', 'positive', 'trust', 'anticipation', 'fear',
       'anger', 'joy', 'surprise', 'sadness']]
```
